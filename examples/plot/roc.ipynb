## Plot ROC Curve (Ungrouped)

```python
argv = ["-i", "data/roc_f1.dat"]

args = roc.parse(argv)

roc.plot(args.input)
```

```python
argv = ["-i", "data/roc_f0.dat", "data/roc_f1.dat", "data/roc_f2.dat",
        "-l", "Fold 1", "Fold 2", "Fold 3"]

args = roc.parse(argv)

roc.plot(args.input, labels=args.labels)
```

## Plot ROC Curve (Grouped)

```python
argv = ["-i", "data/roc_f0.dat", "data/roc_f1.dat", "data/roc_f2.dat",
        "-l", "Group 0", "Group 0", "Group 1",
        "-g", "0", "0", "1"]

args = roc.parse(argv)

roc.plot(args.input, groups=args.groups, labels=args.labels)
```
