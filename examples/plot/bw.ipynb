## Boxplot (without Hue)

A simple boxplot can be plotted from a `pd.DataFrame` by specifying the column representing $x$ (discrete vakues) and the column representing $y$ (continuous values).

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot"
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```

The actual points can be superimposed to the box and whisker plot as a swarm.

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-s",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot and Swarm"
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```

Notches, representing the confidence interval for the median, can be easily added.

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-n",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot"
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```

## Boxplot (with Hue)

Data in the `pd.DataFrame` can be easily grouped together by specifying the hue.

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-g", "smoker",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot (Grouped)",
    "-ln", "Smoker",
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```

The actual points can be superimposed to the box and whisker plot as a swarm. However, the legend is only plotted for the boxplot (and not the swarmplot, which has the same colors).

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-g", "smoker",
    "-s",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot and Swarmplot (Grouped)",
    "-ln", "Smoker",
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```

Notches, representing the confidence interval for the median, can be easily added.

```python
plt.figure(figsize=(10,8))

argv = [
    "-i", "data/tips.csv",
    "-x", "day",
    "-y", "total_bill",
    "-g", "time",
    "-s",
    "-n",
    "-lx", "Day",
    "-ly", "Total Bill ($)",
    "-t", "Boxplot and Swarmplot (Grouped)",
    "-ln", "Smoker",
]

args = bw.parse(argv)

args_dict = bw.args_to_dict(args)

bw.plot(**args_dict)
```
