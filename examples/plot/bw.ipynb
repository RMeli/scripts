```python
import seaborn as sns
from matplotlib import pyplot as plt

tips = sns.load_dataset("tips")
```

## Boxplot (without Hue)

A simple boxplot can be plotted from a `pd.DataFrame` by specifying the column representing $x$ (discrete vakues) and the column representing $y$ (continuous values).

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill", 
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot"
)
```

The actual points can be superimposed to the box and whisker plot as a swarm.

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill",
    swarm=True,
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot and Swarm"
)
```

Notches, representing the confidence interval for the median, can be easily added.

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill",
    notch=True,
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot"
)
```

## Boxplot (with Hue)

Data in the `pd.DataFrame` can be easily grouped together by specifying the hue.

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill", 
    hue_name="smoker", 
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot and Swarmplot",
    legend_name="Smoker",
)
```

The actual points can be superimposed to the box and whisker plot as a swarm. However, the legend is only plotted for the boxplot (and not the swarmplot, which has the same colors).

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill", 
    hue_name="smoker", 
    swarm=True, 
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot and Swarmplot",
    legend_name="Smoker",
)
```

Notches, representing the confidence interval for the median, can be easily added.

```python
plt.figure(figsize=(10,8))

bw.plot(
    df=tips, 
    x_name="day", 
    y_name="total_bill", 
    hue_name="time", 
    swarm=True,
    notch=True,
    xlabel="Day", 
    ylabel="Total Bill ($)", 
    title="Boxplot and Swarmplot"
)
```
